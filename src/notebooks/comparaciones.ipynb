```python
# Importing Libs
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter


%matplotlib inline
```

```python
# Reseting figure size params.
plt.rcParams["figure.figsize"] = [5, 10]

# Setting seaborn as default plotting lib.
sns.set()

sns.set_palette("pastel")

df = pd.read_csv("./central_df.csv")
```

```python
df.shape
```

- Estudio comparativo entre 'algunas' de las categorias de la encuesta.

# TIPO DE CAMINO
- Tipo de camino en el lugar o zona de residencia del encuestado.

```python
df.columns
```

```python
# Agrupacion de tipos de caminos.
tipo_caminos_groups = df.groupby("tipo_camino")
```

# Tipo de camino vs. estado de las calles. [estado_calles]


```python
for k, v in enumerate(tipo_caminos_groups.groups.keys()):
    mod = tipo_caminos_groups.get_group(v)["estado_calles"].value_counts(normalize=True)

    print("Tipo de camino: {0} \n Estado de las calles.".format(v))
    print("=" * 15)
    print(mod)
    print("\n" * 2)
```

```python
plt.figure(figsize=(20, 50), dpi=80)


for k, v in enumerate(tipo_caminos_groups.groups.keys()):
    plt.subplot(8, 2, k + 1)
    mod = (
        tipo_caminos_groups.get_group(v)["estado_calles"]
        .value_counts(normalize=True)
        .plot(
            kind="pie",
            title="Tipo de camino: {0} \n Estado de las calles.".format(v),
            ylabel="",
            autopct="%1.1f%%",
        )
    )
    for l, m in enumerate(
        tipo_caminos_groups.get_group(v)["estado_calles"].value_counts().values
    ):
        mod.annotate(m, (m, l), va="center", color="blue")


plt.subplots_adjust(left=0.6, bottom=0.1, right=1, top=0.7, wspace=0.3, hspace=0.6)
```

    En todas las categorias de caminos puede observarse que los encuestados calificaron, con un porcentaje superior al 40%, como 'LAMENTABLE' el estado de las calles.
