# HELPER FUNCTIONS 

```python
def split_row(row):
    if(isinstance(row, str)):
        return(row.split(';'))
```

```python
def make_df(df, col):
    cats = {}
    for i in col.map(split_row):
        for j in i:
            cats[j]=0
    for i in col.map(split_row):
        for j in i:
            cats[j]+=1
    df = pd.DataFrame(data = [i for i in cats.items()], columns=['cat', 'count']).set_index('cat')
    return(df)
```

# EDUCACION

    - Sobre la 'Educacion' de las personas que respondieron la encuesta.

### 'educación'

```python
df['educación'].value_counts()
```

```python
fig = plt.figure(figsize=(9,5))


nv_ed = make_df(df, df['educación']).plot(kind='barh', title='Nivel Educativo - GENERAL', color=color_neutro_light)

for k,v in enumerate([i for i in make_df(df, df['educación'])['count']]):
       nv_ed.annotate(v,(v,k),  va='center',color='blue')

plt.show()
```

### 'carreras_o_especialidades'

```python
df[ 'carreras_o_especialidades'].value_counts()
```

```python
fig = plt.figure(figsize=(9,5))


carr_df = make_df(df, df[ 'carreras_o_especialidades']).plot(kind='barh', title='Carreras o Especialidades - GENERAL',color=color_neutro_light)

for k,v in enumerate([i for i in make_df(df, df[ 'carreras_o_especialidades'])['count']]):
       carr_df.annotate(v,(v,k),  va='center',color='blue')

plt.show()

```
