### Replacing column values.

```python
central_df = df.loc[df["lugar_residencia_py"].isin(["Central", "Distrito Capital"])]
```

```python
rep_val = {"Distrito Capital": "Central"}
```

```python
aux_f = central_df.replace({"lugar_residencia_py": rep_val})
central_df = aux_f
```
