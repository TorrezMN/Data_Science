# Estado de las calles.
    - Se les pregunto a los encuestados su consideracion sobre el estado de las calles.

```python
estado_calles_group = df.groupby("estado_calles")
```

```python
estado_calles_group.groups.keys()
```

## Tipo de camino
- Se evalua el estado de las calles segun el "tipo de camino".

```python
plt.figure(figsize=(20, 30), dpi=80)


for k, v in enumerate(estado_calles_group.groups.keys()):
    plt.subplot(8, 2, k + 1)
    mod = (
        estado_calles_group.get_group(v)["tipo_camino"]
        .value_counts(normalize=True)
        .plot(
            kind="pie",
            title="Estado de las calles: {0} \n Tipo de camino.".format(v),
            ylabel="",
            autopct="%1.1f%%",
        )
    )
    for l, m in enumerate(
        estado_calles_group.get_group(v)["tipo_camino"].value_counts().values
    ):
        mod.annotate(m, (m, l), va="center", color="blue")


plt.subplots_adjust(left=0.6, bottom=0.1, right=1, top=0.9, wspace=0.3, hspace=0.6)
```

## Tipo de camino
- Tiempo que demoran las reparaciones.

```python
plt.figure(figsize=(15, 30), dpi=80)


for k, v in enumerate(estado_calles_group.groups.keys()):
    plt.subplot(8, 2, k + 1)
    mod = (
        estado_calles_group.get_group(v)["tiempo_reparaciones"]
        .value_counts()
        .plot(
            kind="barh",
            title="Estado de las calles: {0} \n Tiempo de reparaciones.".format(v),
        )
    )
    for l, m in enumerate(
        estado_calles_group.get_group(v)["tiempo_reparaciones"].value_counts().values
    ):
        mod.annotate(m, (m, l), va="center", color="blue")


plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=1)
```

## Sufrio accidentes

## Regularidad de las reparaciones
- Regularidad de las reparaciones viales.

```python
plt.figure(figsize=(10, 30), dpi=80)


for k, v in enumerate(estado_calles_group.groups.keys()):
    plt.subplot(8, 2, k + 1)
    mod = (
        estado_calles_group.get_group(v)["regularidad_reparaciones"]
        .value_counts()
        .plot(
            kind="pie",
            title="Estado de las calles: {0} \n Regularidad de las reparaciones.".format(
                v
            ),
            ylabel="",
            autopct="%1.1f%%",
        )
    )
    for l, m in enumerate(
        estado_calles_group.get_group(v)["regularidad_reparaciones"]
        .value_counts()
        .values
    ):
        mod.annotate(m, (m, l), va="center", color="blue")


plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.5)
```
